```python
# Please test the code with as model is already saved.......................

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense

# Initialising the CNN
classifier = Sequential() # It creates a pipeline............


'''32--Kernel Filter
3, 3--Size of the Filter
64, 64, 3--Image size
'relu'-Acticvation Function'''

# Step 1 - Convolution
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
classifier.add(Conv2D(32, (3, 3),  activation = 'relu'))
classifier.add(Conv2D(32, (3, 3),  activation = 'relu'))
classifier.add(Conv2D(32, (3, 3),  activation = 'relu'))
classifier.add(Conv2D(32, (3, 3),  activation = 'relu'))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer




# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full connection
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 128, activation = 'relu'))

classifier.add(Dense(units = 4, activation = 'softmax')) # Changing This

# Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy') # Changing This loss = 'categorical_crossentropy'

# Part 2 - Fitting the CNN to the images

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)

'''C:/Users/shrey/Ineuron/DL/CNN/Images_'''

training_set = train_datagen.flow_from_directory('C:/Users/shrey/Ineuron/DL/CNN/Images_New',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode='binary')

test_set = test_datagen.flow_from_directory('C:/Users/shrey/Ineuron/DL/CNN/Images_New',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode='binary')

model = classifier.fit_generator(training_set,
                         steps_per_epoch = 8000,
                         epochs = 1,
                         validation_data = test_set,    
                         validation_steps = 2000)


'''
In single epoch it is going to send=8000*128 of the data

--- (100xxxx(Total Data)/8000)=y
--- Now, there are y cycles in each epoch. Hence, total there are 8000 cycles(1 epoch).
'''





classifier.save("model.h6")
print("Saved model to disk")


# Note: That target_size = (64, 64)) and  input_shape = (64, 64, 3) size must be same...................
```
